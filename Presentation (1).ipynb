## Matplotlib

Matplotlib is a Python 2D plotting library which can be used in Python scripts, the Python and IPython shells, the Jupyter notebook, web application servers, and four graphical user interface toolkits.
Matplotlib is the most popular plotting library for python.
It gives you control over every aspect of a figure.
It was designed to have a similar feel to Matlab's graphical plotting.


Some examples are - Line plot, Histogram, Scatter plot, 3D plot, Image plot, Contour plot, Scatter plot, Polar plot, Lineplot, 3-D plot & Image plot

Let us look at a very simple plot. 
First let's see the necessary modules that has to be imported.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns #Importing pandas to create a dataframe that will help us access the features easily.
%matplotlib inline    
```

Now, let's do some basic plotting

```python
# Prepare the data
x = np.linspace(0, 10, 100)

# Plot the data
plt.plot(x, x, label='linear')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

We can also change the style of the graph. To plot a bar graph, following code is used.


```python
from matplotlib import style
plt.style.use('ggplot')  #Pre-defined style

#'Solarize_Light2', '_classic_test_patch', 'bmh', 'classic', 'dark_background', 'fast', 
#'fivethirtyeight', 'ggplot', 'grayscale', 'seaborn', 'seaborn-bright', 'seaborn-colorblind', 'seaborn-dark',
#'seaborn-dark-palette', 'seaborn-darkgrid', 'seaborn-deep', 'seaborn-muted', 'seaborn-notebook', 'seaborn-paper',
#'seaborn-pastel', 'seaborn-poster', 'seaborn-talk', 'seaborn-ticks', 'seaborn-white', 'seaborn-whitegrid', 
#'tableau-colorblind10'

x = [1,2,3,4]
y = [1,2,3,4]



plt.bar(x, y)  #Bar graph


plt.title('Info')  #To set title
plt.ylabel('Y axis')    #To name the X-axis
plt.xlabel('X axis')    #To name the y-axis

plt.show()
```

Let's us look at a histogram. The code is -

```python

plt.bar([1,3,5,7,9],[5,2,7,8,2], label="Example one")

plt.bar([2,4,6,8,10],[8,6,2,5,6], label="Example two", color='g')  #We can change colors using color parameter
plt.legend()
plt.xlabel('bar number')
plt.ylabel('bar height')

plt.title('Bar\nGraph')

plt.show()
```

We can also specify the bins in the histogram, using bins parameter.

You can also change their width using rwidth. 

You can use alpha here for engaging the transperancy in the bars, this is usually used when we have to visualize multiple histograms based on different visualizations on the data for x-axis.

Orientation can be used to rotate your bars/histograms

```python

population_ages = [22,55,62,45,21,22,34,42,42,4,99,102,110,120,121,122,130,111,115,112,80,75,65,54,44,43,42,48]


plt.hist(population_ages, bins=5, histtype='bar',color="#0504aa",alpha=0.5, rwidth=0.8,
        orientation='horizontal')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Histogram With Only\nOnly 5 Bins')
plt.show()
```

Here's an example of a scatter plot.

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

plt.scatter(x,y, label='skitscat', color='k', s=25, marker="o")  #Marker allows you to specify the type, try using v
                                                                 #s = size of points

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
```

Her's a pie plot

```python


slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']
 
plt.pie(slices,
  labels=activities,   
  colors=cols,   #Specify colors for each feature
  startangle=90,  #Specify the rotation and how you want the plotting to be done 
  shadow= True,
  explode=(0,0.1,0,0),  #Seperation of the slices
  autopct='%1.1f%%')     #Showing the percentages of each slice 
plt.title('Pie Plot')
plt.show()
```

# Bokeh

Bokeh is an interactive visualization library in Python.
What sets it apart from other visualization libraries is the that
it targets web browsers for representation.



Some advantages for Bokeh are :-

 1) Build Complex statistical plots
 
 2) Provides visualizations in various outputs like html,server
 
 3) Helpful in converting visualizations written in other libraries
   like matplotlib, seaborn ,ggplot
   
 4) Flexibilty factors like applying interaraction, layout styling etc.

```python
from bokeh.io import output_file, show
from bokeh.plotting import figure
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
output_file('line.html')
fig = figure(title = 'Line Plot example', x_axis_label = 'x', y_axis_label = 'y')
fig.line(x,y)  #Line graph
show(fig)
```

```python
fig = figure(plot_width = 400, plot_height = 200)
fig.hbar(y = [2,4,6], height = 1, left =1, right = [1,2,3], color = "Blue")
output_file('bar.html')
show(fig)
```

```python
import random
p = figure()
s=[]
s1=[]
for i in range(100):
    s.append(random.randint(0,500))
    s1.append(random.randint(0,500))
tools = "hover, box_zoom, crosshair"
p = figure(tools=tools)
p.scatter(s, s1, marker = "circle", size = 10, fill_color = "NAVY")
output_file('scatter.html')
show(p)
```

```python

fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']
counts = [54, 32, 42, 45, 44, 6]

p = figure(x_range=fruits, plot_height=250, title="Fruit Counts", tools="")

p.vbar(x=fruits, top=counts,color='green',alpha=0.6, width=0.9)


show(p)
```

```python
animals =  ['Tiger', 'Lion', 'Bear', 'Rhino', 'Leopard', 'Cheetah']
years = ["2015", "2016", "2017"]
colors = ["#c9d9d3", "#718dbf", "#e84d60"]

data = {'animals' : animals,
        '2015'   : [212, 122, 433, 101, 222, 301],
        '2016'   : [200, 132, 411, 87, 291,342],
        '2017'   : [187, 204, 409, 200, 275, 355]}

p = figure(x_range=animals, plot_height=250, title="Animal Counts by Year",
           toolbar_location=None, tools="hover", tooltips="$name @animals: @$name")

p.vbar_stack(years, x='animals', width=0.9, color=colors, source=data)


show(p)




```

```python
sample1 = np.random.multivariate_normal([0,0], [[1,0],[0,1]], 5000)
sample2 = np.random.multivariate_normal([1,1], [[0.05,0],[0,0.05]], 5000)
sample = np.concatenate([sample1, sample2])
tools = "hover, box_zoom, undo, crosshair"
p = figure(tools=tools)
p.scatter(sample[:,0], sample[:,1], alpha=0.5)
show(p)
```
